```python
#2. Crie uma função recursiva para calcular o n-ésimo número da sequência de Fibonacci.

def fibonacci(x, a = 0, b = 1, c = 1):
    if(c == x):
        return b
    else:
        a, b = b, a+b
        c += 1
        return fibonacci(x, a, b, c)
n = int(input(f'Digite a quantidade termos da Sequencia: '))
fibonacci(n)
```

```python
# 11. Escreva uma função que verifique se duas palavras são anagramas
# (mesmas letras em ordem diferente). 
# Ex: 
#  amor – Roma = True;
#  casa – caro = False

def ver_anagrama(palavra1,palavra2):
    if(len(palavra1)==len(palavra2)):
        
        palavra1 = sorted(palavra1)
        palavra2 = sorted(palavra2)

        if(palavra1==palavra2):
            print(f"As palavras digitadas são anagramas!!")
        else:
            print(f"Não é um anagrama")
    else:
        print(f"Não é um anagrama")
        
p1 = input("Digite a primeira palavra")
p2 = input("Digite a segunda palavra para ser verificada:")

ver_anagrama(p1,p2)
```

```python
# 12. Crie uma função que simule uma calculadora com operações
# básicas (soma, subtração, multiplicação, divisão) 

def somar(a,b):
    a = a+b
    return a

def multiplicar(a,b):
    a = a*b
    return a

def subtrair(a,b):
    a = a-b
    return a

def dividir(a,b):
    a = a/b
    return a
def remove(listar,i):
    listar.pop(i+1)
    listar.pop(i-1)
    return listar

def calcula(listac):
    for i in range(len(listac)):
        if(listac[i] == "*"):
            listac[i] = multiplicar(listac[i+1],listac[i-1])
            remove(listac,i)
            return calcula(listac)
        elif(listac[i] == "/"):
            listac[i] = dividir(listac[i-1],listac[i+1])
            remove(listac,i)
            return calcula(listac)
    for i in range(len(listac)):
        if(listac[i] == "+"):
            listac[i] = somar(listac[i-1],listac[i+1])
            remove(listac,i)
            return calcula(listac)
        elif(listac[i] == "-"):
            listac[i] = subtrair(listac[i-1],listac[i+1])
            remove(listac,i)
            return calcula(listac)
        
    return listac

lista2 = []
while(True):
    numeros = int(input("Digite um numero:"))
    lista2.append(numeros)
    numeros = input("Digite o operador:\n * -> Multiplicar\n  + -> Somar\n - -> Subtrair\n / -> Dividir\n = -> Resolver Equação")
    if(numeros == "="):
        break
    lista2.append(numeros)
calcula(lista2)
```

```python
# 14. Crie uma função recursiva que calcule a soma dos números de 1 até n.


soma = 0
def somar_numeros(valor):
    global soma
    if (valor != 0):
        soma += valor
        valor-=1
        return somar_numeros(valor)
    else:
        return soma

n = int(input("Até qual número deseja realizar as somas ?"))
somar_numeros(n)
```

```python
# 17. Crie uma função recursiva que receba um número inteiro positivo e 
# retorne a soma de seus dígitos. Exemplo: soma_digitos(123) -> 6 (1 + 
# 2 + 3) 

soma = 0
def somar_digitos(a):
    global soma
    if(a != 0 ):
        soma += a%10
        a //= 10
        return somar_digitos(a)
    return soma

n = int(input("Digite o valor para somarmos os digitos"))
somar_digitos(n)

```
