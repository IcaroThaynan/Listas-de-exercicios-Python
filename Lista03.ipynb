```python
# Questão 06 - Escreva um algoritmo que calcule o fatorial de um número digitado pelo usuário. Na
# matemática, o fatorial de um número natural n, denotado por n!, é o produto de todos os
# naturais menores ou iguais a n.
# Fórmula: n! = n × (n - 1)! × ... × 1

numero = int(input("Digite um número para calcular o fatorial: "))
cont = numero
while(cont > 1 ):
    cont-=1
    numero = numero * cont
print(f"O resultado é: {numero}")
```

```python
# Questão 08 - Escreva um algoritmo que verifique se um determinado número digitado pelo usuário
# é primo. Números primos são todos os números naturais que possuem somente dois
# divisores: 1 e eles mesmos.

numero = int(input("Digite um número para verificar se é primo: "))
resultado = 0
controle = 1

while(controle <= numero):
    if(numero%controle == 0): #condição de resto 0
        resultado+=1 #soma quantas vezes entrou na condição
    controle+=1

if(resultado == 2): #Teste de quantas vezes entrou na condição
    print("Número primo!!")
else:
    print("Não é primo!")
```

```python
#Questão 13 - Escreva um algoritmo que leia 10 números e mostre a média, mediana e a moda dos
# números digitados.

numeros = []
cont = 0
soma = 0
while(cont < 10):
    n = int(input("Digite um número: "))
    soma+=n
    numeros.append(n)
    cont+=1
print(f"Média: {soma/10}")
numeros = sorted(numeros) # ordena a lista
soma = (numeros[4] + numeros[5]) / 2 #calcula mediana
print(f"A mediana é: {soma}")

```

```python
#Questão 15 - Escreva um algoritmo que receba dois valores um que representa o inicio e outro o
# fim. Em seguida imprima uma contagem regressiva personalizada.

numero1 = int(input("Digite um número "))
numero2 = int(input("Digite o outro número: "))

if ( numero1 > numero2):
    maior = numero1
    menor = numero2
else:
    maior = numero2
    menor = numero1

while(maior >= menor):
    print(f"{maior}")
    maior-=1
```

```python
#Questão 16 -Escreva um algortimo que verifique se um número é palíndromo e imprima uma
# mensagem em cada caso. Uma palavra ou número palíndromo é aquele cuja sequência de
# letras/números é simétrica, permitindo uma leitura idêntica da esquerda para a direita ou
# da direita para a esquerda: ovo, osso, reler, anilina.

n = int(input("Digite um numero para verificar se é palindromo:"))
b = str()
c = str(n)

while (n // 10 > 0):
    b += str(n%10)
    n = int(n/10)
b += str(n%10)
print(f"{b}")
if(b == c):
    print(f"{c} É um Palindromo")
else:
    print(f"{c} Não é um Palíndromo!")
```
